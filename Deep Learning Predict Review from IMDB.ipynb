```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/drive/MyDrive/Mini_project data science/IMDB Dataset.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
# แปลงค่าคอลัมน์ 'sentiment' เป็น 1 และ 0 โดยใช้ np.where
df['sentiment'] = np.where(df['sentiment']== 'positive',1,0)
```

```python
df.head()
```

```python
df['sentiment'].unique()
```

```python
print("Total Class Positive :",(df['sentiment']== 1).sum())
print("Total Class Negative :",(df['sentiment']== 0).sum())
```

```python
x = df['review']
y = df['sentiment']
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
# Tokenize ข้อความ
tokenizer = Tokenizer(num_words=5000, oov_token="<OOV>")
tokenizer.fit_on_texts(x)
sequences = tokenizer.texts_to_sequences(x)
# Pad sequences ให้มีความยาวเท่ากัน
max_len = 200  # กำหนดความยาวสูงสุด
X_padded = pad_sequences(sequences, maxlen=max_len, padding='post', truncating='post')
```

```python
# Train test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_padded, y, test_size=0.2, random_state=42)
```

```python
# สร้างโมเดล
model = Sequential([
    Embedding(input_dim=5000, output_dim=128, input_length=max_len),
    LSTM(64, return_sequences=False),
    Dense(1, activation='sigmoid')
])
```

```python
# คอมไพล์และฝึกโมเดล
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()

model.fit(X_train, y_train, epochs=10, batch_size=16, validation_data=(X_test, y_test))
```

```python
# ทำนายผล
predictions = model.predict(X_test)
```

```python
print(predictions[:])
```

```python
from sklearn.metrics import accuracy_score

# คำนวณ accuracy
accuracy = accuracy_score(y_test, predictions.round())
print(f"Accuracy: {accuracy:.2f}")
```

```python
new_reviews = [
    "I really enjoyed this movie, it was fantastic!",
    "This film was awful, a complete waste of time.",
    "The movie was okay, not great but not bad either.",
    "I loved the characters and the plot was so interesting!",
    "The film was very slow and boring, I didn't enjoy it at all.",
    "It was a masterpiece, I would recommend it to anyone!",
    "This was the worst movie I've ever seen. So disappointing.",
    "Amazing storyline, incredible acting. Definitely worth watching.",
    "I didn't like the movie, it felt too predictable.",
    "The movie was average. Nothing special, but not terrible."
]
```

```python
# แปลงข้อความใหม่เป็นลำดับตัวเลขโดยใช้ Tokenizer ที่ฝึกไว้
new_sequences = tokenizer.texts_to_sequences(new_reviews)
```

```python
# เติมเต็มหรือย่อความยาวให้เท่ากับข้อมูลที่ใช้ฝึก
new_padded = pad_sequences(new_sequences, maxlen=max_len, padding='post', truncating='post')
```

```python
# ทำนายผลจากข้อความใหม่
predictions = model.predict(new_padded)

# แปลงผลลัพธ์เป็นคลาส (0 หรือ 1) โดยใช้เกณฑ์ 0.5
predicted_classes = (predictions >= 0.5).astype(int)

# แสดงผลการทำนาย
for review, pred_class in zip(new_reviews, predicted_classes):
    sentiment = 'Positive' if pred_class == 1 else 'Negative'
    print(f"Review: {review}\nSentiment: {sentiment}\n")
```
